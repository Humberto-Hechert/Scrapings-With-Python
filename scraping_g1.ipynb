```python
#Importando as Bibliotecas
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Pegando as informações da portal de tecnologia do g1
url = 'https://g1.globo.com/tecnologia/'
response = requests.get(url)
site = BeautifulSoup(response.text, 'html.parser')

noticias = site.findAll('div', {'class': 'bastian-feed-item'})
lista = []
for noticia in noticias:
  item = {}
  item['titulo'] = noticia.find('a', {'class': 'feed-post-link gui-color-primary gui-color-hover'}).get_text()
  item['descricao'] = noticia.find('div', {'class': 'feed-post-body-resumo'}).get_text()
  item['imagem'] = noticia.find('img', {'class': 'bstn-fd-picture-image'})['src']
  item['link'] = noticia.find('a', {'class': 'feed-post-figure-link gui-image-hover'})['href']

  lista.append(item)
lista
```

```python
#Jogando as noticias em um Data Frame
df = pd.DataFrame(lista)
df
```

```python
#Transportando para csv
df.to_csv('scraping_g1.csv', sep=';', index = False, encoding = 'utf-8')
```
