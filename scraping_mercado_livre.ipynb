```python
#Importando as Bibliotecas
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Pegando as informações necessárias para o scraping
url_base = 'https://lista.mercadolivre.com.br/'

#Aqui você escolhe qual produto deseja escolher, utilizamos 'ps5' como exemplo
produto_nome = input('Qual produto deseja? ')

response = requests.get(url_base + produto_nome)

site = BeautifulSoup(response.text, 'html.parser')

produtos = site.findAll('div', {'class': 'ui-search-result__wrapper'})

lista = []

for produto in produtos:
  item = {}
  titulo = produto.find('h2', {'class': 'ui-search-item__title ui-search-item__group__element'})

  link = produto.find('a', {'class': 'ui-search-result__content ui-search-link'})

  preco_simbolo = produto.find('span', {'class': 'price-tag-symbol'})
  preco_valor = produto.find('span', {'class': 'price-tag-fraction'})

  item['titulo'] = titulo.text
  item['link'] = link['href']
  item['preco'] = (preco_simbolo.text + preco_valor.text)

  lista.append(item)
```

```python
#Imprimindo o resultado do scraping
lista
```

```python
#Jogando o resultado em um Data Frame
df = pd.DataFrame(lista)
df
```

```python
#Exportando os dados para um arquivo csv
df.to_csv('scraping_mercado_livre.csv', sep=';', index = False, encoding = 'utf-8')
```
